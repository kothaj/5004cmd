```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

# load both datasets
td = pd.read_csv("Trips_by_Distance.csv", parse_dates=["Date"])
tf = pd.read_csv("Trips_Full Data.csv", parse_dates=["Date"])

# keep only national level data
td = td[td["Level"] == "National"]
tf = tf[tf["Level"] == "National"]

# merge the required columns using Date
model_data = pd.merge(
    tf[["Date", "Trips 1-25 Miles"]],
    td[["Date", "Number of Trips 5-10"]],
    on="Date",
    how="inner"
)

# remove missing values
model_data = model_data.dropna()

# define x and y
x = model_data[["Trips 1-25 Miles"]]
y = model_data[["Number of Trips 5-10"]]

print("x shape:", x.shape, ", y shape:", y.shape)

# split into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.3, random_state=42
)

# -------------------------
# Linear Regression
# -------------------------
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
y_pred_linear = linear_model.predict(x_test)

# -------------------------
# Polynomial Regression
# -------------------------
poly = PolynomialFeatures(degree=2)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)

poly_model = LinearRegression()
poly_model.fit(x_train_poly, y_train)
y_pred_poly = poly_model.predict(x_test_poly)

# -------------------------
# Print Results
# -------------------------
print("\nLinear Regression Results")
print(f"R² Score: {r2_score(y_test, y_pred_linear):.4f}")
print(f"MSE: {mean_squared_error(y_test, y_pred_linear):.2f}")
print(f"Intercept: {linear_model.intercept_[0]:.2f}")
print(f"Coefficients: {linear_model.coef_[0][0]:.2f}")

print("\nPolynomial Regression Results")
print(f"R² Score: {r2_score(y_test, y_pred_poly):.4f}")
print(f"MSE: {mean_squared_error(y_test, y_pred_poly):.2f}")
print(f"Intercept: {poly_model.intercept_[0]:.2f}")
print(f"Coefficients: {poly_model.coef_.flatten().tolist()}")
```
