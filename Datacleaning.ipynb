```python
import dask.dataframe as dd

# fix mixed data type columns
dtype = {
    "County Name": "object",
    "State Postal Code": "object"
}

# load datasets using dask
df = dd.read_csv(
    "Trips_by_Distance.csv",
    parse_dates=["Date"],
    assume_missing=True,
    dtype=dtype
)

df_full = dd.read_csv(
    "Trips_Full Data.csv",
    parse_dates=["Date"],
    assume_missing=True
)

# check missing values
missing_values = df.isna().sum().compute()
print("Missing values in Trips_by_Distance.csv:")
print(missing_values)
```

```python

missing_full = df_full.isna().sum().compute()
print("\nMissing values in Trips_Full Data.csv:")
print(missing_full)

# fill missing values in one important column
df["Population Staying at Home"] = df["Population Staying at Home"].fillna(
    df["Population Staying at Home"].mean()
)
```

```python

# check duplicates
df_pd = df.compute()
duplicates = df_pd.duplicated().sum()
print("\nDuplicate rows in Trips_by_Distance.csv:", duplicates)
```
