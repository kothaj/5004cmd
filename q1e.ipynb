```python
import pandas as pd
import matplotlib.pyplot as plt

# load dataset
tf = pd.read_csv("Trips_Full Data.csv", parse_dates=["Date"])

# keep only national level data
tf = tf[tf["Level"] == "National"]

# calculate mean travellers by distance
distance_means = tf[[
    "Trips 1-25 Miles",
    "Trips 25-50 Miles",
    "Trips 50-100 Miles",
    "Trips 100-250 Miles",
    "Trips 250-500 Miles",
    "Trips 500+ Miles"
]].mean()

print(distance_means)

# plot
plt.figure(figsize=(10, 5))
distance_means.plot(kind="bar")
plt.xlabel("Trip Distance Category")
plt.ylabel("Average Number of Travellers")
plt.title("Average Travellers by Trip Distance")
plt.show()
```
