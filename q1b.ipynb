```python
import pandas as pd
import matplotlib.pyplot as plt

# load dataset
td = pd.read_csv("Trips_by_Distance.csv", parse_dates=["Date"])

# keep only national level data
td = td[td["Level"] == "National"]

# -------------------------
# 1. 10-25 range above 10 million
# -------------------------
dates_10_25 = td[td["Number of Trips 10-25"] > 10000000][["Date", "Number of Trips 10-25"]]

plt.figure(figsize=(10, 5))
plt.scatter(dates_10_25["Date"], dates_10_25["Number of Trips 10-25"])
plt.xlabel("Date")
plt.ylabel("Number of Trips 10-25")
plt.title("Dates with More Than 10,000,000 Trips in 10-25 Range")
plt.show()

# -------------------------
# 2. 50-100 range above 10 million
# -------------------------
dates_50_100 = td[td["Number of Trips 50-100"] > 10000000][["Date", "Number of Trips 50-100"]]

plt.figure(figsize=(10, 5))
plt.scatter(dates_50_100["Date"], dates_50_100["Number of Trips 50-100"])
plt.xlabel("Date")
plt.ylabel("Number of Trips 50-100")
plt.title("Dates with More Than 10,000,000 Trips in 50-100 Range")
plt.show()
```
